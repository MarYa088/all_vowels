```python
""" ЗАДАНИЕ к теме "Условия и циклы": 
    Написать функцию number_of_vowels() - подсчет количества гласных в тексте.
Принимает - текст.
Возвращает - кол-во гласных (русских, английских), суммированное из обоих регистров (верх. и нижнего).
Выводит - словарь вида: {буква: кол-во вхождений}.
Требуемое решение - необходимо написать код без использования библиотечной функции 
                    перевода текста в единый регистр.
"""
""" АЛГОРИТМ
1. Создать упорядоченный лист гласных - vowels: 
 [строчные англ, заглавные англ(в том же порядке), строчные рус, заглавные рус(в том же порядке)]
2. Посчитать количество вхождений в текст каждого символа из vowels:
   - создать словарь all_big_small_vowels{символ: количество вхождений}

Если в тексте нет гласных в верхнем регистре - работа сделана, можно будет вернуть all_big_small_vowels

3. Проверить, есть ли в all_big_small_vowels гласные из верхнего регистра. Для этого использовать 
  свойство vowels - фиксированную разность между позициями строчной и соответствующей ей заглавной.
  В английских символах разность = 6, расположены с 0 по 5 (строчные) и с 6 по 11 (заглавные).
  В русских символах разность = 10, расположены с 12 по 21 (строчные) и с 22 по 31 (заглавные).
   - Создать словарь position_in_vowels{номер_позиции: символ}
   - Посчитать, сколько раз каждый символ входит в текст,
     создать словарь number_of_positions_in_txt{номер_позиции: количество_вхождений_в_текст}
   - Создать словари sum_upper_and_lower_eng/rus 
     {номер_буквы_англ/рус: кол-во вхождений для тех, для которых были вхождения заглавных}
     Для этого найти совпадающие символы в all_big_small_vowels и vowels[6:12][22:32]
     и посчитать количество вхождений каждого.
     Если к ним в тексте есть парные строчные - прибавить кол-во заглавных к кол-ву строчных.
     Если к ним в тексте нет парных строчных - просто присвоить кол-во заглавных - парной строчной.
     
     Если бы в Питоне можно было напрямую совершать математические операции с позициями ключей 
     словарей, алгоритм был бы проще.

4. Дополнить словари sum_upper_and_lower_eng/rus недостающими гласными, 
   т.е. теми, которые имеются только в нижнем регистре.
   - Создать словари total_number_of_eng/rus 
     {номер_буквы_англ/рус: итоговое кол-во вхождений по всему тексту}
     
5. Создать словарь total_number{номер_буквы: итоговое кол-во вхождений по всему тексту}.
   Для этого сконкатенировать total_number_of_eng и total_number_of_rus
   
6. Создать итоговый словарь all_vowels{буква: итоговое кол-во вхождений по всему тексту}

Работа сделана!
"""
```

```python
def number_of_vowels(txt):
    vowels = (['a','e','i','o','u','y','A','E','I','O','U','Y',
               'а','о','и','е','ё','э','ы','у','ю','я','А','О','И','Е','Ё','Э','Ы','У','Ю','Я'])

# all_big_small_vowels (кол-во вхождений в текст каждого символа из vowels)
# pre_vowel_count (ф-я считает кол-во вхождений в текст каждого символа из vowels)
    all_big_small_vowels = pre_vowel_count(txt, vowels)

# position_in_vowels (Позиция буквы в vowels), number_of_positions_in_txt (количество позиций в тексте)
    position_in_vowels, number_of_positions_in_txt = (
        number_of_positions_in_text(all_big_small_vowels, vowels)
    )
#    print('number_of_positions_in_txt',number_of_positions_in_txt,'\n------')
      
    sum_upper_and_lower_eng, sum_upper_and_lower_rus = (
        sum_upper_and_lower(all_big_small_vowels, vowels, position_in_vowels, number_of_positions_in_txt)
    )
    
    
    ''' проверяем, есть ли гласные в верхнем регистре:
        НЕТ англ, НЕТ рус:
    '''
    if len(sum_upper_and_lower_rus) == 0 and len(sum_upper_and_lower_eng) == 0:
        return all_big_small_vowels
    
    ''' ЕСТЬ англ, ЕСТЬ рус:
    '''
    if len(sum_upper_and_lower_rus) > 0 and len(sum_upper_and_lower_eng) > 0:
        total_number_of_eng = total_sum_eng(sum_upper_and_lower_eng, number_of_positions_in_txt)

        total_number_of_rus = total_sum_rus(sum_upper_and_lower_rus, number_of_positions_in_txt)

    ''' ЕСТЬ англ, НЕТ рус:
    '''
    if len(sum_upper_and_lower_eng) > 0 and len(sum_upper_and_lower_rus) == 0:
        total_number_of_eng = total_sum_eng(sum_upper_and_lower_eng, number_of_positions_in_txt)

        total_number_of_rus = no_upper_rus(number_of_positions_in_txt, sum_upper_and_lower_rus)

    ''' НЕТ англ, ЕСТЬ рус:
    '''
    if len(sum_upper_and_lower_eng) == 0 and len(sum_upper_and_lower_rus) > 0:
        total_number_of_rus = total_sum_rus(sum_upper_and_lower_rus, number_of_positions_in_txt)

        total_number_of_eng = no_upper_eng(number_of_positions_in_txt, sum_upper_and_lower_eng)

    total_number = {** total_number_of_eng, **total_number_of_rus} # объединенный словарь

    all_vowels = vowels_all(total_number, position_in_vowels)
    return all_vowels
```

```python
def pre_vowel_count(txt, vowels):
    '''
    считает: кол-во вхождений в текст каждого символа из vowels
    принимает: - текст
                - лист гласных (vowels)
    возвращает: all_big_small_vowels словарь кол-ва вхождений в текст символов из переменной vowels
    '''
    
    all_big_small_vowels = {}
    for char_vowels in vowels:
        for char_txt in txt:
            if char_vowels == char_txt:
                if char_vowels in all_big_small_vowels:
                    all_big_small_vowels[char_vowels] += 1
                else:
                    all_big_small_vowels[char_vowels] = 1
                    
    return all_big_small_vowels
```

```python
def number_of_positions_in_text(all_big_small_vowels, vowels):
    '''
    принимает:
        - словарь all_big_small_vowels кол-ва вхождений в текст символов из переменной vowels
        - лист гласных (vowels)
    возвращает:
        - position_in_vowels (словарь номера позиции в vowels)
        - number_of_positions_in_txt (словарь номеров символов с кол-вом вхождений каждого из них)
    '''
    
    position_in_vowels = {} # словарь номера позиции в vowels
    s = -1#0
    for charact in vowels:
        s += 1
        position_in_vowels[s] = charact
#    print('position_in_vowels ',position_in_vowels,'\n--------')
    
    number_of_positions_in_txt = {} # словарь номеров символов с кол-вом вхождений каждого из них
    for character in all_big_small_vowels:
        for char in position_in_vowels:
            if character == position_in_vowels[char]:
                number_of_positions_in_txt[char] = all_big_small_vowels[character]
                
    return position_in_vowels, number_of_positions_in_txt
  
```

```python
def sum_upper_and_lower(all_big_small_vowels, vowels, position_in_vowels, number_of_positions_in_txt):
    '''
    принимает:
        - словарь all_big_small_vowels кол-ва вхождений в текст символов из переменной vowels
        - position_in_vowels (словарь номера позиции в vowels)
        - number_of_positions_in_txt (словарь номеров символов с кол-вом вхождений каждого из них)
        - лист гласных (vowels)
    возвращает:
        - sum_upper_and_lower_eng (словарь- номер_буквы_англ: кол-во вхождений для тех, для которых были
        вхождения заглавных)
        - sum_upper_and_lower_rus (словарь- номер_буквы_рус: кол-во вхождений для тех, для которых были
        вхождения заглавных)
    '''  

    # словарь номер_буквы_анг: кол-во вхождений(с нумерацией только по строчным независимо от регис):
    sum_upper_and_lower_eng = {}
    
    for i in all_big_small_vowels:
        for j in vowels[6:12]: # все заглавные англ
            if i == j: # нашлась заглавная буква
                # выясняем, есть ли в all_big_small_vowels такая строчная
                # сначала выясняем ее номер (=k[0]) по словарю position_in_vowels=({1: 'a', 2: 'e', 3: 'i'...):
                for k in list(position_in_vowels.items()):
                    if k[1] == i:
                        # проверяем, есть ли в (number_of_positions_in_txt =all_big_small_vowels с номерами вместо букв) ключ = (k[0]-6)
                        if k[0]-6 in number_of_positions_in_txt: # если есть - увеличиваем кол-во строчных на кол-во заглавн
                            sum_upper_and_lower_eng[k[0]-6] = number_of_positions_in_txt[k[0]-6] + number_of_positions_in_txt[k[0]]
                        else: # если нет - присваиваем номеру строчной кол-во заглавных
                            sum_upper_and_lower_eng[k[0]-6] = number_of_positions_in_txt[k[0]]
            # УРА,я нашла заглавные буквы англ.
            # Если к ним уже были парные строчные - прибавила кол-во заглавных с кол-ву строчных
            # Если к ним не было парных строчных - просто присвоила кол-во заглавных - парной строчной
    #print('sum_upper_and_lower_eng ',sum_upper_and_lower_eng,'\n------')
            
    # словарь номер_буквы_рус: кол-во вхождений(с нумерацией только по строчным независимо от регис):
    sum_upper_and_lower_rus = {}
            
    for i in all_big_small_vowels:
        for j in vowels[22:32]: # все заглавные рус
            if i == j: # нашлась заглавная буква
                # выясняем, есть ли в all_big_small_vowels такая строчная
                # сначала выясняем ее номер (=k[0]) по словарю position_in_vowels=({1:'a', 2:'e', 3:'i'...):
                for k in list(position_in_vowels.items()):
                    if k[1] == i:
                        # проверяем, есть ли в (number_of_positions_in_txt =all_big_small_vowels с номерами вместо букв) ключ = (k[0]-6)
                        if k[0]-10 in number_of_positions_in_txt: # если есть - увеличиваем кол-во строчных на кол-во заглавн
                            sum_upper_and_lower_rus[k[0]-10] = number_of_positions_in_txt[k[0]-10] + number_of_positions_in_txt[k[0]]
                        else: # если нет - присваиваем номеру строчной кол-во заглавных
                            sum_upper_and_lower_rus[k[0]-10] = number_of_positions_in_txt[k[0]]
            # УРА,я нашла заглавные буквы рус.
            # Если к ним уже были парные строчные - прибавила кол-во заглавных с кол-ву строчных
            # Если к ним не было парных строчных - просто присвоила кол-во заглавных - парной строчной
    #print('sum_upper_and_lower_rus ',sum_upper_and_lower_rus,'\n------')

    return sum_upper_and_lower_eng, sum_upper_and_lower_rus
```

```python
def total_sum_eng(sum_upper_and_lower_eng, number_of_positions_in_txt):
    '''
    принимает:
        - sum_upper_and_lower_eng (словарь- номер_буквы_англ: кол-во вхождений для тех, для которых были
        вхождения заглавных)
        - number_of_positions_in_txt (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_number_of_eng (словарь- номер_буквы_англ: итоговое кол-во вхождений по всему тексту)
    '''  
    total_number_of_eng = sum_upper_and_lower_eng # словарь - итоговое кол-во вхождений по всему тексту
    
    for i in number_of_positions_in_txt:
        if i < 6:
            if (i in sum_upper_and_lower_eng) == False:
                total_number_of_eng[i] = number_of_positions_in_txt[i]
                #print('total_number_of_eng ',total_number_of_eng)

    return total_number_of_eng
  
```

```python
def total_sum_rus(sum_upper_and_lower_rus, number_of_positions_in_txt):
    '''
    принимает:
        - sum_upper_and_lower_rus (словарь- номер_буквы_рус: кол-во вхождений для тех, для которых были
        вхождения заглавных)
        - number_of_positions_in_txt (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_number_of_rus (словарь- номер_буквы_рус: итоговое кол-во вхождений по всему тексту)
    '''  
    total_number_of_rus = sum_upper_and_lower_rus # словарь рус вхождений независимо от регистра
    
    for i in number_of_positions_in_txt:
        if i > 11 and i < 22:
            if (i in sum_upper_and_lower_rus) == False:
                total_number_of_rus[i] = number_of_positions_in_txt[i]
                #print('total_number_of_rus ',total_number_of_rus)
    
    return total_number_of_rus
```

```python
def vowels_all(total_number, position_in_vowels):
    '''
    принимает:
        - total_number (словарь- номер_буквы: кол-во вхождений)
        - position_in_vowels (словарь номера позиции в vowels)
    возвращает:
        - all_vowels (словарь- итоговое кол-во вхождений по всему тексту: буква)
    '''      
    all_vowels = {}
    for i in total_number:
        for j in position_in_vowels:
            if i == j:
                all_vowels[position_in_vowels[i]] = total_number[i]
                break
                
    return all_vowels
```

```python
# нет рус заглавных
def no_upper_rus(number_of_positions_in_txt, sum_upper_and_lower_rus):
    '''
    принимает:
        - sum_upper_and_lower_rus (словарь- номер_буквы_рус: кол-во вхождений для тех, для которых были
        вхождения заглавных)        
        - number_of_positions_in_txt (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_number_of_rus (словарь - номер_буквы_рус: кол-во вхождений для каждой независимо от регис)
    ''' 
    total_number_of_rus = {}
    for i in number_of_positions_in_txt:
        if i > 11:
            total_number_of_rus[i] = number_of_positions_in_txt[i]
        
    return total_number_of_rus
```

```python
# нет англ заглавных
def no_upper_eng(number_of_positions_in_txt, sum_upper_and_lower_eng):
    ''' 
    принимает:
        - sum_upper_and_lower_eng (словарь- номер_буквы_англ: кол-во вхождений для тех, для которых были
        вхождения заглавных)        
        - number_of_positions_in_txt (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_number_of_eng (словарь - номер_буквы_англ: кол-во вхождений для каждой независимо от регис)
    ''' 
    total_number_of_eng = {}
    for i in number_of_positions_in_txt:
        if i < 6:
            total_number_of_eng[i] = number_of_positions_in_txt[i]
        
    return total_number_of_eng
```
