```python
def function(txt):
    vowels = (['a','e','i','o','u','y','A','E','I','O','U','Y',
               'а','о','и','е','ё','э','ы','у','ю','я','А','О','И','Е','Ё','Э','Ы','У','Ю','Я'])
#    if 'vowels' in locals():
#        print('Variable exist.')    

    total = pre_total(txt, vowels)
    
    pos = numbering(total, vowels)[0]
    posn = numbering(total, vowels)[1]
      
    nb_total_eng = eng_to_lower(total, vowels, pos, posn)        
    nb_total_rus = rus_to_lower(total, vowels, pos, posn)
    
    ''' проверяем, есть ли гласные в верхнем регистре:
        НЕТ англ, НЕТ рус:
    '''
    if len(nb_total_rus) == 0 and len(nb_total_eng) == 0:
        return total
    ''' ЕСТЬ англ, ЕСТЬ рус:
    '''
    if len(nb_total_rus) > 0 and len(nb_total_eng) > 0:
        total_eng_nonb = is_eng(nb_total_eng, posn)
        total_rus_nonb = is_rus(nb_total_rus, posn)                   
    ''' ЕСТЬ англ, НЕТ рус:
    '''
    if len(nb_total_eng) > 0 and len(nb_total_rus) == 0:
        total_eng_nonb = is_eng(nb_total_eng, posn)
        total_rus_nonb = no_rus(posn, nb_total_rus)  
    ''' НЕТ англ, ЕСТЬ рус:
    '''
    if len(nb_total_eng) == 0 and len(nb_total_rus) > 0:
        total_rus_nonb = is_rus(nb_total_rus, posn)
        total_eng_nonb = no_eng(posn, nb_total_eng)
   
    total_nonb = {** total_eng_nonb, **total_rus_nonb} # объединенный словарь   
    all_vowels = vowels_all(total_nonb, pos)
    return all_vowels
```

```python
def pre_total(txt, vowels):
    '''
    принимает: - текст
            - лист гласных (vowels)
    возвращает: словарь кол-ва вхождений в текст символов из переменной vowels
    '''
    
    nmbr_of_occ = {} # словарь количества вхождений для каждой буквы
    for k in vowels:
        nmbr_of_occ[k] = 0
    for i in txt:
        if i in vowels:
            nmbr_of_occ[i] += 1
            
    total = {} # словарь количества вхождений >0
    for i in nmbr_of_occ.keys():
        if nmbr_of_occ[i] != 0:
            total[i] = nmbr_of_occ[i]

    return total
```

```python
def numbering(total, vowels):
    '''
    принимает:
        - словарь total кол-ва вхождений в текст символов из переменной vowels
        - лист гласных (vowels)
    возвращает:
        - pos (словарь номера позиции в vowels)
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
    '''
    
    pos = {} # словарь номера позиции в vowels
    s = 0
    for j in vowels:
        s += 1
        pos[s] = j
    
    posn = {} # словарь номеров символов с кол-вом вхождений каждого из них
    for i in total:
        for j in pos:
            if i == pos[j]:
                posn[j] = total[i]
                
    return pos, posn
  
```

```python
def eng_to_lower(total, vowels, pos, posn):
    '''
    принимает:
        - словарь total кол-ва вхождений в текст символов из переменной vowels
        - pos (словарь номера позиции в vowels)
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
        - лист гласных (vowels)
    возвращает:
        - nb_total_eng (словарь- номер_буквы_англ: кол-во вхождений для тех, для которых были
        вхождения заглавных)
    '''  

    # словарь номер_буквы_анг: кол-во вхождений(с нумерацией только по строчным независимо от регис):
    nb_total_eng = {}
    
    for i in total:
        for j in vowels[6:12]: # все заглавные англ
            if i == j: # нашлась заглавная буква
                # выясняем, есть ли в total такая строчная
                # сначала выясняем ее номер (=k[0]) по словарю pos=({1: 'a', 2: 'e', 3: 'i'...):
                for k in list(pos.items()):
                    if k[1] == i:
                        # проверяем, есть ли в (posn =total с номерами вместо букв) ключ = (k[0]-6)
                        if k[0]-6 in posn: # если есть - увеличиваем кол-во строчных на кол-во заглавн
                            nb_total_eng[k[0]-6] = posn[k[0]-6] + posn[k[0]]
                        else: # если нет - присваиваем номеру строчной кол-во заглавных
                            nb_total_eng[k[0]-6] = posn[k[0]]
            # УРА,я нашла заглавные буквы англ.
            # Если к ним уже были парные строчные - прибавила кол-во заглавных с кол-ву строчных
            # Если к ним не было парных строчных - просто присвоила кол-во заглавных - парной строчной
    return nb_total_eng
```

```python
def rus_to_lower(total, vowels, pos, posn):
    '''
    принимает:
        - словарь total кол-ва вхождений в текст символов из переменной vowels
        - pos (словарь номера позиции в vowels)
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
        - лист гласных (vowels)
    возвращает:
        - nb_total_rus (словарь- номер_буквы_рус: кол-во вхождений для тех, для которых были
        вхождения заглавных)
    '''  

    # словарь номер_буквы_рус: кол-во вхождений(с нумерацией только по строчным независимо от регис):
    nb_total_rus = {}
            
    for i in total:
        for j in vowels[22:32]: # все заглавные рус
            if i == j: # нашлась заглавная буква
                # выясняем, есть ли в total такая строчная
                # сначала выясняем ее номер (=k[0]) по словарю pos=({1: 'a', 2: 'e', 3: 'i'...):
                for k in list(pos.items()):
                    if k[1] == i:
                        # проверяем, есть ли в (posn =total с номерами вместо букв) ключ = (k[0]-6)
                        if k[0]-10 in posn: # если есть - увеличиваем кол-во строчных на кол-во заглавн
                            nb_total_rus[k[0]-10] = posn[k[0]-10] + posn[k[0]]
                        else: # если нет - присваиваем номеру строчной кол-во заглавных
                            nb_total_rus[k[0]-10] = posn[k[0]]
            # УРА,я нашла заглавные буквы рус.
            # Если к ним уже были парные строчные - прибавила кол-во заглавных с кол-ву строчных
            # Если к ним не было парных строчных - просто присвоила кол-во заглавных - парной строчной

    return nb_total_rus
```

```python
def is_eng(nb_total_eng, posn):
    '''
    принимает:
        - nb_total_eng (словарь- номер_буквы_англ: кол-во вхождений для тех, для которых были
        вхождения заглавных)
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_eng_nonb (словарь- номер_буквы_англ: итоговое кол-во вхождений по всему тексту)
    '''  
    total_eng_nonb = nb_total_eng # словарь - итоговое кол-во вхождений по всему тексту
    
    for i in posn:
        if i > 6:
            break
        for j in nb_total_eng:
            if i == j:
                break
            else:
                total_eng_nonb[i] = posn[i]
                break
                
    return total_eng_nonb
  
```

```python
def is_rus(nb_total_rus, posn):
    '''
    принимает:
        - nb_total_rus (словарь- номер_буквы_рус: кол-во вхождений для тех, для которых были
        вхождения заглавных)
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_rus_nonb (словарь- номер_буквы_рус: итоговое кол-во вхождений по всему тексту)
    '''  
    total_rus_nonb = nb_total_rus # словарь рус вхождений независимо от регистра
        
    for i in posn:
        if i > 11 and i < 22:
            for j in nb_total_rus:
                if i == j:
                    break
                else:
                    total_rus_nonb[i] = posn[i]
                    break
                        
    return total_rus_nonb
```

```python
def vowels_all(total_nonb, pos):
    '''
    принимает:
        - total_nonb (словарь- номер_буквы: кол-во вхождений)
        - pos (словарь номера позиции в vowels)
    возвращает:
        - all_vowels (словарь- итоговое кол-во вхождений по всему тексту: буква)
    '''      
    all_vowels = {}
    for i in total_nonb:
        for j in pos:
            if i == j:
                all_vowels[pos[i]] = total_nonb[i]
                break
                
    return all_vowels
```

```python
# нет рус заглавных
def no_rus(posn, nb_total_rus):
    '''
    принимает:
        - nb_total_rus (словарь- номер_буквы_рус: кол-во вхождений для тех, для которых были
        вхождения заглавных)        
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_rus_nonb (словарь - номер_буквы_рус: кол-во вхождений для каждой независимо от регис)
    ''' 
    total_rus_nonb = {}
    for i in posn:
        if i > 12:
            total_rus_nonb[i] = posn[i]
        
    return total_rus_nonb
```

```python
# нет англ заглавных
def no_eng(posn, nb_total_eng):
    '''
    принимает:
        - nb_total_eng (словарь- номер_буквы_англ: кол-во вхождений для тех, для которых были
        вхождения заглавных)        
        - posn (словарь номеров символов с кол-вом вхождений каждого из них)
    возвращает:
        - total_eng_nonb (словарь - номер_буквы_англ: кол-во вхождений для каждой независимо от регис)
    ''' 
    total_eng_nonb = {}
    for i in posn:
        if i < 7:
            total_eng_nonb[i] = posn[i]
        
    return total_eng_nonb
```
